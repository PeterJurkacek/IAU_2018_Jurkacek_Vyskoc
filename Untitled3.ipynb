```python
class Abc(TransformerMixin):
    def __init__(self, column, source, target):
        self.column = column
        self.source = source
        self.target = target
        
    def fit(self, df, y=None, **fit_params):
        return self
    
    def transform(self, df, **transform_params):
        sample = df[self.column] == self.source
        df.loc[sample, self.column] = self.target
        return df
```

```python
eu_countries = ["Hungary","Poland","Ireland","France","England","Portugal","Germany","Italy","Holand-Netherlands","Greece"]

def translate(country):
    if country in eu_countries:
        country = "Europe"
    elif country == "Unknown country":
        return country
    elif country != "United-States":
        country = "Other"
    return country
```

### Date of birth


Dátumy majú rôzny formát- je potrebné ho upraviť

```python
def correct_date_format(date):
    date = str(date).replace('/', '-')
    date = date[:10]
    return date 
    
```

```python
data['date_of_birth'] = data['date_of_birth'].map(correct_date_format)
#data['date_of_birth'] = pd.to_datetime(data['date_of_birth'], format = '%Y-%m-%d %H:%M:%S')
#data['date_of_birth'] = data['date_of_birth'].dt.strftime('%m/%d/%Y')
```

```python
def date(text):
    year = text.split("-")
    if year[0] != 'nan':
        if len(year[0]) != 4:
            if len(year[2]) == 2 and int(year[0]) > 31:
                new_date = "19"+year[0]+"-"+year[1]+"-"+year[2] 
            elif len(year[2]) == 2 and int(year[0]) < 31 and int(year[2]) > 31:
                new_date = "19"+year[2]+"-"+year[1]+"-"+year[0] 
            else:
                new_date = year[2]+"-"+year[1]+"-"+year[0] 
            return new_date
    return text
```

```python
data['date_of_birth'] = data['date_of_birth'].map(date)
```

```python
import statsmodels.api as smm
%pylab inline

y = data.T3
x = data.TT4
x = smm.add_constant(x)
x.head()

est = smm.OLS(y,x)

est = est.fit()

est.summary()
est.params
X_prime = np.linspace(x.TT4.min(), x.TT4.max(), 100)[:, np.newaxis]
X_prime = sm.add_constant(X_prime) # add constant as we did before
y_hat = est.predict(X_prime)
plt.scatter(x.TT4, y, alpha=0.3) # Plot the raw data
plt.xlabel("TT4")
plt.ylabel("T3")
plt.plot(X_prime[:, 1], y_hat, 'r', alpha=0.9) # Add the regression line, colored in red

```

```python
# import formula api as alias smf

import statsmodels.formula.api as smf
# formula: response ~ predictors
est = smf.ols(formula='T3 ~ TT4', data=data).fit()
est.summary()
```

```python
# Fit the no-intercept model

est_no_int = smf.ols(formula='T3 ~ TT4 - 1', data=data).fit()
# We pick 100 hundred points equally spaced from the min to the max
X_prime_1 = pd.DataFrame({'TT4': np.linspace(x.TT4.min(), x.TT4.max(), 100)})
X_prime_1 = sm.add_constant(X_prime_1) # add constant as we did before

y_hat_int = est.predict(X_prime_1)
y_hat_no_int = est_no_int.predict(X_prime_1)

fig = plt.figure(figsize=(8,4))
splt = plt.subplot(121)

splt.scatter(x.TT4, y, alpha=0.3) # Plot the raw data
plt.ylim(30, 100) # Set the y-axis to be the same
plt.xlabel("TT4")
plt.ylabel("T3")
plt.title("With intercept")
splt.plot(X_prime[:, 1], y_hat_int, 'r', alpha=0.9) # Add the regression line, colored in red

splt = plt.subplot(122)
splt.scatter(x.TT4, y, alpha=0.3) # Plot the raw data
plt.xlabel("TT4")
plt.title("Without intercept")
splt.plot(X_prime[:, 1], y_hat_no_int, 'r', alpha=0.9) # Add the regression line, colored in red
```
